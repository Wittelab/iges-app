```python
from fuzzywuzzy import fuzz
from matplotlib_venn import venn2
import matplotlib.pyplot as plt
```

## ChatGTP takes a stab

```python
# Sample lists of abstract titles
list1 = [line.rstrip() for line in open('IGES abstracts.txt', 'r')]
list2 = [line.rstrip() for line in open('ASHG abstracts.txt', 'r')]

print(f"# Of IGES abstract titles: {len(list1):,}")
print(f"# Of ASHG abstract titles: {len(list2):,}")
```

```python
# Define a similarity threshold (adjust as needed)
threshold = 80

# Find similar titles using fuzzy string matching
similar_titles = set()
for title1 in list1:
    for title2 in list2:
        if fuzz.token_sort_ratio(title1, title2) >= threshold:
            similar_titles.add(title1)
            similar_titles.add(title2)

# Create a Venn diagram to visualize the results
venn2([set(list1), set(list2)], set_labels=("List 1", "List 2"))
plt.title("Fuzzy Matching of Abstract Titles")
plt.show()

# Display the similar titles
print("Similar Abstract Titles:")
for title in similar_titles:
    print(title)
```

**NOTE: Using set() in the venn2 command defeats the purpose of fuzzy matching, since set matching now resorts back to exact matches only!**

## A more human approach

```python
import pandas as pd
```

```python
similar = pd.DataFrame(columns=['Threshold', 'Score', 'IGES','ASHG'])
```

```python
for threshold in range(50,100,10):
    print(f"Looking at threshold {threshold}")
    # Find similar titles using fuzzy string matching
    similar_titles = set()
    for title1 in list1:
        for title2 in list2:
            score = fuzz.token_sort_ratio(title1, title2)
            if  score >= threshold:
                similar.loc[len(similar.index)] = [threshold, score, title1, title2] 
```

```python
# Do this in another loop so we can replot quickly 
for threshold in range(50,100,10):
    df = similar[similar['Threshold']==threshold].drop_duplicates(subset='IGES')
    # Create a Venn diagram to visualize the results
    # subsets=(size A, size B, size A and B)
    venn2(subsets = (len(list1)-df.shape[0], len(list2)-df.shape[0], df.shape[0]), 
          set_labels = ('IGES', 'ASHG'),
          set_colors=('purple', 'skyblue'), alpha = 0.7)
    plt.title(f"Fuzzy Matching of Abstracts at Match Threshold {threshold}")
    plt.show()
```

```python
similar[['Score','IGES','ASHG']]\
    .sort_values('Score', ascending=False)\
    .drop_duplicates()\
    .to_csv("Similar abstracts.tsv", header=True, sep="\t", index=False)
```
